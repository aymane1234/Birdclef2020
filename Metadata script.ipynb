# Script for preparing the Meta data 

Description: The Metadata file contains 960 folders named after the Bird species, each folder contains json files which we want to get into a Dataframe shape by the following script:


```python
import json
import os
import pandas as pd
```

```python
def getJSON(filePathAndName):
    with open(filePathAndName,'r') as fp:
        return json.load(fp)
```

```python
Json_fpath = "./metadata/Acanthis flammea_Common Redpoll/"   ##jason-file Path (1st kind --->Acanthis flammea_Common Redpoll)
Json_files = os.listdir(Json_fpath)
print("No. of .JSON files in the folder = ",len(Json_files))
```

```python
Data = getJSON(Json_fpath+Json_files[0])
Data
```

# Construct a dataframe from the rows( As above) generated by each json file in the respective folder

```python
rows = []
for i in range(0,len(Json_files)-1):
    data = pd.DataFrame(getJSON(Json_fpath+Json_files[i]))
    rows.append(data)
    
df_Acanthis_flammea = pd.concat(rows,axis = 0)
```

```python
df_Acanthis_flammea   ###An example of a dataframe for a respictive bird kind
```

# Now we generate a dataframe for all bird kinds

```python
meta_SAMPLE_DIR = "./metadata/"     ##directory where the json_files are stored (root directory)
```

```python
sample_fullpaths = []               ###Get a list of paths

for root, dirs, files in os.walk(meta_SAMPLE_DIR):
    for file in files:
        if file.endswith(".json"):
            sample_fullpaths.append(os.path.join(root, file))
            
```

```python
cols = []
for filename in sample_fullpaths:
    data = pd.DataFrame(getJSON(filename))
    cols.append(data) ####get a list of dataframs(rows)

data_final = pd.concat(cols,axis = 0)  ###concatenate to get a big dataframe
    
```

```python
data_final 
```

#  Store Metadata


```python
data_final.to_csv( './metadata/data_final.csv') 
```

# Preparing the Audio files

```python
# directory where the audio samples are stored (root directory)
AUD_SAMPLE_DIR                      = "./audio/"
```

```python
sample_fullpaths = []
sample_track_ids = []

for root, dirs, files in os.walk(AUD_SAMPLE_DIR):
    for file in files:
        if file.endswith(".mp3"):
            sample_fullpaths.append(os.path.join(root, file))
            sample_track_ids.append(file.replace(".mp3", ""))

metadata_audio          = pd.DataFrame(sample_fullpaths, index=sample_track_ids, columns=["audio_path"])
metadata_audio["audio"] = True

print("%d AUD audio samples found" % metadata_audio.shape[0])
```
