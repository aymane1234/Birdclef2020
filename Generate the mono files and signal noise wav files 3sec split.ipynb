# This part deals with changing the stereo to mono channel if the files have already only one channel skip this to finally use the pr.preprocess_sound_file

```python
from pydub import AudioSegment
```

```python
for wav_file in filenames:
    sound = AudioSegment.from_wav(wav_file)
    sound = sound.set_channels(1)
    sound.export(wav_file[0:-4] + '_mono' + '.wav', format="wav")
    print(wav_file + ' is converted')
```
